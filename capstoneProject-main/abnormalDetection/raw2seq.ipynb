```python
import os
import numpy as np
```

```python
actions = ['error', 'suffer', 'falling', 'lying', 'sitting', 'walking', 'standing', 'lain', 'jump']
seq_length = 30
```

```python
filePath = "dataset"
data = []
idx = 0

os.makedirs('dataset', exist_ok=True)

for (root, directories, files) in os.walk(filePath):
    for file in files:
        if file.startswith("seq"):# exclude seq dataset
            continue
        print(file)
        file_path = os.path.join(root, file)
        data = np.load(file_path)
        if data.shape[0] == 0:# exclude unavailable array
            continue
        action = actions[int(data[0, -1])]

        # Create sequence data
        full_seq_data = []
        for seq in range(len(data) - seq_length):
            full_seq_data.append(data[seq:seq + seq_length]) # increase axis by sliding window(seq_length)

        full_seq_data = np.array(full_seq_data)
        print(action, full_seq_data.shape)
        np.save(os.path.join('dataset', f'seq_{action}_{idx}'), full_seq_data)

        idx += 1
```
